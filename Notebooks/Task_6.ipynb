```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Parquet_to_CSV") \
    .getOrCreate()

# Load Parquet dataset
df = spark.read.parquet("nifty_10m_fixed.parquet")

# Save as a single CSV file
df.coalesce(1) \
  .write \
  .mode("overwrite") \
  .option("header", True) \
  .csv("nifty_10m_single_csv")
```

```python
import shutil

shutil.make_archive(
    "nifty_10m_single_csv",
    "zip",
    "nifty_10m_single_csv"
)
```
