```python
from pyspark.sql import SparkSession
import pandas as pd
import os
from datasets import load_dataset
```

## Downloading Dataset

```python
from datasets import load_dataset
import pandas as pd

ds = load_dataset("artist-23/nifty-options-data")

sample_10m = ds["train"].select(range(10000000))

pdf = sample_10m.to_pandas()

# Convert timestamp columns to string
pdf["timestamp"] = pdf["timestamp"].astype(str)
pdf["datetime"] = pdf["datetime"].astype(str)
pdf["date"] = pdf["date"].astype(str)

pdf.to_parquet("nifty_10m_fixed.parquet", index=False)
```

## Creating Spark Session


```python
spark = SparkSession.builder \
    .appName("NiftyOptionsAnalysis") \
    .getOrCreate()
```

## Loading dataset


```python
df = spark.read.parquet("nifty_10m_fixed.parquet")
```

## Displaying records


```python
df.show(20)
```

## Schema


```python
df.printSchema()
```

## Counting Rows and Columns

```python
# Row count
print("Rows:", df.count())
# Column count
print("Columns:", len(df.columns))
```

## File size


```python
size_bytes = os.path.getsize("nifty_10m_fixed.parquet")
size_mb = size_bytes / (1024 * 1024)
size_gb = size_bytes / (1024 * 1024 * 1024)

print(f"File Size: {size_mb:.2f} MB")
print(f"File Size: {size_gb:.2f} GB")
```

## Saving the file as Parquet


```python
df.write \
  .mode("overwrite") \
  .parquet("nifty_10m_processed.parquet")
```
